```python
#논리연산자
#1. and 연산자

print(True and True)
print(True and False)
print(False and False)
print(True and True and False)

# 2. or 연산자
print(True or True)
print(True or False)
print(False or False)

# 3. not 연산자
print(not True)
print(not False)
print(not 1)
print(not [])
```
