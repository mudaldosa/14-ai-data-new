```python
!pip install matplotlib
# 1️⃣ 라이브러리 불러오기
import matplotlib.pyplot as plt

# 2️⃣ 데이터 준비
fruits = ['apple', 'banana', 'cherry', 'date']
counts = [10, 15, 7, 12]

# 3️⃣ 막대그래프 그리기
plt.bar(fruits, counts, color='skyblue')  # 막대 색 지정
plt.title('Fruit Counts')                 # 그래프 제목
plt.xlabel('Fruit')                       # x축 라벨
plt.ylabel('Count')                       # y축 라벨

# 4️⃣ 그래프 보여주기
plt.show()
```

```python
# 딕셔너리 쌍 추가하기
a = {1: "a"}
a[2]="b"
print(a)
a["name"] = "ken"

print(a)
a[(1, 2)] = "hobby"
print(a)
a[[3,4]] = "job"
print(a)
```

```python
# 딕셔너리 요소 삭제하기
print(a)
del a[1]
print(a)

del a["name"]
print(a)
```

```python
# 딕셔너리에서 key를 활용해 value 얻기
profile ={
    "name" : "홍길동", 
    "age" : 25
}
print(profile["name"])
```

```python
#딕셔너리에 관련된 함수
# 1. key리스트 만들기 - keys
profile = {
    "name" : "홍길동", 
    "age" : 25,
    "번호" : "010-1234-1234"
}
print(profile.keys())

# 2. value 리스트 만들기 - values
profile = {
    "name" : "홍길동", 
    "age" : 25,
    "번호" : "010-1234-1234"
}
print(profile.values())

# 3. key:value 쌍 얻기 - items
profile = {
    "name" : "홍길동", 
    "age" : 25,
    "번호" : "010-1234-1234"
}
print(profile.items())
print(profile)

# 4. key, value 쌍 모두 지우기 - clear
profile = {
    "name" : "홍길동", 
    "age" : 25,
    "번호" : "010-1234-1234"
}
print(profile.clear())

# 5. key로 value 얻기 -get
profile = {
    "name" : "홍길동", 
    "age" : 25,
    "번호" : "010-1234-1234"
}
print(profile.get("name"))

# 해당 key가 딕셔너리 안에 있는지 조사하기 - in
profile = {
    "name" : "홍길동", 
    "age" : 25,
    "번호" : "010-1234-1234"
}
print("name" in profile)

# key로 value 얻기 - pop

profile = {
    "name" : "홍길동", 
    "age" : 25,
    "번호" : "010-1234-1234"
}
temp = profile.pop("name")
print(temp)
print(profile)
```
