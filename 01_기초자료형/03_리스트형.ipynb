```python
#리스트 연산하기
# 1. 리스트 더하기
a =[1,2,3]
b =[4,5,6]
print(a + b)

# 2. 리스트 곱하기(반복)
a =[1,2,3]
print(a * 3)
```

```python
# 리스트의 수정과 삭제
# 1. 리스트의 값 수정하기

a = [1, 2, 3]
a[2] = 4
print(a)

# 2. 리스트의 요소 삭제하기
a = [1, 2, 3]
del a[1]
print(a)

a = [1, 2, 3, 4, 5]
del a[2:]
```
